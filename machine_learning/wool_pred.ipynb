```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
categorical_columns = ['Crimp Characteristics', 'Strength', 'Elasticity', 'Fineness']

for column in categorical_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

# Encode the target variable
le_target = LabelEncoder()
df['Wool Type'] = le_target.fit_transform(df['Wool Type'])
print("Mapping of Wool Type classes:", dict(zip(le_target.classes_, range(len(le_target.classes_)))))
print("Encoded Wool Type values:", df['Wool Type'].unique())

X=df.drop('Wool Type',  axis=1)
y=df['Wool Type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
print("\nEncoded DataFrame:\n", df.head(50))
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print("\nEncoded DataFrame:\n", df.head())
```

```python
# using decision tree model no 1

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python

# using xgb classifier model no 2

# train
import xgboost as xgb
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
model.fit(X_train, y_train)

# pred
y_pred = model.predict(X_test)

# accuracy% check
from sklearn.metrics import classification_report, accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
#Mapping of Wool Type classes: {'Coarse Wool': 0, 'Fine Wool': 1, 'Medium Wool': 2}
```

```python
# using Support Vector Machine (SVM) with Linear Kernel model no 3

from sklearn.svm import SVC
model = SVC(kernel='linear', random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pickle
with open('trained_model2.pkl', 'wb') as f:
    pickle.dump(model, f)
```

```python
with open('label_encoders.pkl', 'wb') as encoder_file:
    pickle.dump(label_encoders, encoder_file)

# Save the target encoder (if you need it for decoding predictions)
with open('target_encoder.pkl', 'wb') as target_encoder_file:
    pickle.dump(le_target, target_encoder_file)
```
